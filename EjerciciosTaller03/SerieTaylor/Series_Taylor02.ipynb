```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Función original
# --------------------------
def original_function(x):
    return 1 / (1 - x)

# --------------------------
# Aproximación de Taylor
# --------------------------
def taylor_approximation(x, order):
    return sum(x**n for n in range(order + 1))

# --------------------------
# Ecuaciones en formato texto
# --------------------------
def get_taylor_equation(order):
    terms = []
    for n in range(order + 1):
        if n == 0:
            terms.append("1")
        else:
            terms.append(f"x^{n}")
    return " + ".join(terms)

# --------------------------
# Dominio y órdenes
# --------------------------
x = np.linspace(-1.5, 0.9, 400)
x_valid = x[x < 1]  # Evitar división por cero en x = 1
orders = list(range(0, 6))  # Órdenes del 0 al 5

# --------------------------
# Crear figura con subplots
# --------------------------
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), gridspec_kw={'height_ratios': [3, 1]})

# --------------------------
# Parte 1: Gráfico
# --------------------------
ax1.set_title('Aproximaciones de Taylor para $\\frac{1}{1 - x}$ alrededor de $x_0 = 0$', fontsize=16)
ax1.plot(x_valid, original_function(x_valid), 'k', linewidth=3, label='$\\frac{1}{1 - x}$')

# Graficar cada polinomio de Taylor
for order in orders:
    y = taylor_approximation(x, order)
    ax1.plot(x, y, '--', linewidth=1.5, label=f'Orden {order}')

# Decoración del gráfico
ax1.axhline(0, color='black', linewidth=0.5)
ax1.axvline(0, color='black', linewidth=0.5)
ax1.axvline(x=1, color='red', linestyle=':', linewidth=2, alpha=0.7, label='Singularidad en $x=1$')

ax1.set_xlabel('x', fontsize=14)
ax1.set_ylabel('y', fontsize=14)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(loc='upper left', fontsize=10)
ax1.set_xlim(-1.5, 1.5)
ax1.set_ylim(-1, 5)

# --------------------------
# Parte 2: Ecuaciones
# --------------------------
ax2.axis('off')  # Oculta los ejes

# Texto con las ecuaciones
equation_text = "Aproximaciones de Taylor de $\\frac{1}{1 - x}$ alrededor de $x_0 = 0$:\n\n"
for order in orders:
    equation = get_taylor_equation(order)
    equation_text += f"Orden {order}:  $P_{{{order}}}(x) = {equation}$\n"

ax2.text(0.01, 1, equation_text, fontsize=12, va='top', ha='left', family='monospace')

# Ajustar diseño
plt.tight_layout()
plt.show()
```
