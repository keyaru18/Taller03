```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Datos originales
x_data = np.array([-2, 1, 3, 7])
y_data = np.array([5, 7, 11, 34])

# Calcular el polinomio de Lagrange
poly = lagrange(x_data, y_data)

# Crear rango para la grfica
x_plot = np.linspace(-3, 8, 500)
y_plot = poly(x_plot)

# Configuracin de la grfica
plt.figure(figsize=(14, 8))
plt.title('Interpolacin de Lagrange con Desarrollo Matemtico', fontsize=16)

# Graficar los polinomios base individuales
colors = ['red', 'green', 'blue', 'purple']
for i in range(4):
    # Crear polinomio base para cada punto
    base_poly = lagrange(x_data, np.eye(4)[i])
    plt.plot(x_plot, base_poly(x_plot), '--', color=colors[i], alpha=0.5, 
             label=f'$L_{i}(x)$ (Base para punto {x_data[i]}, {y_data[i]})')

# Graficar el polinomio final
plt.plot(x_plot, y_plot, 'k-', linewidth=3, label='Polinomio de Lagrange $P(x)$')

# Graficar puntos originales
plt.plot(x_data, y_data, 'ro', markersize=8, label='Puntos dados')

# Mostrar el polinomio en el grfico
polynomial_text = (f"$P(x) = {poly.coeffs[0]:.4f}x^3 + {poly.coeffs[1]:.4f}x^2 + "
                  f"{poly.coeffs[2]:.4f}x + {poly.coeffs[3]:.4f}$")
plt.text(0.02, 0.98, polynomial_text, transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.8))

# Configuraciones adicionales
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.xlim(-3, 8)
plt.ylim(0, 40)

# Mostrar tabla con los clculos
col_labels = ['Punto', 'x_i', 'y_i', 'Polinomio Base L_i(x)']
cell_text = [
    ['0', '-2', '5', '$\\frac{(x-1)(x-3)(x-7)}{-135}$'],
    ['1', '1', '7', '$\\frac{(x+2)(x-3)(x-7)}{36}$'],
    ['2', '3', '11', '$-\\frac{(x+2)(x-1)(x-7)}{40}$'],
    ['3', '7', '34', '$\\frac{(x+2)(x-1)(x-3)}{216}$']
]

plt.table(cellText=cell_text, colLabels=col_labels,
          loc='bottom', bbox=[0.1, -0.5, 0.8, 0.4])

plt.tight_layout()
plt.subplots_adjust(bottom=0.3)
plt.show()
```
